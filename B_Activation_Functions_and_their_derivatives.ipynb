### Activation functions and its derivatives

<ul>
    <li>Sigmoid activaton function</li>
    <li>Tanh activaton function</li>
    <li>ReLU activaton function</li>
    <li>Leaky ReLU activaton function</li>
    
</ul>

<p><b>ReLu</b> - Rectified Linear unit is the default choice of activation functions in the <b>hidden layer</b>. 
In the <b>output layer</b>, we use <b>Sigmoid</b> as activation function, because its output is in the range between 0 and 1.</p>
<p>
    By using ReLu in the hidden layer, the Neural Network will <b>learn much faster</b> then using sigmoid or tanah, becasue the slope of sigmoid and tanh is going to be 0 if z is large positive or negative number and it slow down gradient descent. The derivative of ReLu is 1 if z>0.
    </p>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
z = np.linspace(-10,10,100)
```

#### Sigmoid activation function

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

# derivative of Sigmoid Function

def dsigmoid(a):
    return a*(1-a) # returns a derivative od sigmoid function if a=sigmoid then a'=a(1-a)
```

```python
# we will now plot sigmoid activation function ant it's derivative
plt.plot(z, sigmoid(z), label = 'sigmoid')
plt.plot(z, dsigmoid(sigmoid(z)), label = r'$ \frac{dsigmoid}{dz}$')
plt.legend(fontsize = 12)
plt.xlabel('z')
plt.show()
```

#### Tanh activation function

```python
def tanh(z):
    return np.tanh(z)

# derivative of tanh

def dtanh(a):
    return 1-np.power(a,2)
```

```python
plt.plot(z, tanh(z),'b', label = 'tanh')
plt.plot(z, dtanh(tanh(z)),'r', label=r'$ \frac{dtanh}{dz}$')
plt.legend(fontsize = 12)
plt.show()
```

#### ReLu activation function

```python
def ReLU(z):
    return np.maximum(0,z)

# derivative of ReLu

def dReLU(a):
    return 1*(a>0)
```

```python
plt.plot(z, ReLU(z),'b', label ='ReLU')
plt.plot(z, dReLU(ReLu(z)),'r', label=r'$ \frac{dReLU}{dz}$')
plt.legend(fontsize = 12)
plt.xlabel('z')
plt.show()
```

#### Leaky ReLU activation function

```python
def LeakyReLU(z):
    return np.maximum(0.01*z,z)

# derivative of ReLu
def dLeakyReLU(a):
    return 0.01*(a>0)
```

```python
plt.plot(z, LeakyReLU(z),'b', label = 'LeakyReLU')
plt.plot(z, dLeakyReLU(LeakyReLU(z)),'r', label=r'$ \frac{dLeakyReLU}{dz}$')
plt.legend(fontsize = 12)
plt.xlabel('z')
plt.show()
```

#### Comparison of derivatives of activation functions

```python
plt.plot(z, dsigmoid(sigmoid(z)),label = r'$ \frac{dsigmoid}{dz}$' )
plt.plot(z, dtanh(tanh(z)), label = r'$ \frac{dtanh}{dz}$')
plt.plot(z, dReLU(ReLU(z)), label=r'$ \frac{dReLU}{dz}$')
plt.plot(z, dLeakyReLU(LeakyReLU(z)), label=r'$ \frac{dLeakyReLU}{dz}$')
plt.legend(fontsize = 12)
plt.xlabel('z')
plt.title('Derivatives of activation functions')
plt.show()
```
